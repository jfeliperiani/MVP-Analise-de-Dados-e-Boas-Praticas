# Descrição do Problema

O conjunto de dados selecionado corresponde a dados reais de internações hospitalares no setor suplementar de saúde (operadoras de planos privados de saúde), disponibilizados pela Agência Nacional de Saúde Suplementar – ANS no formato de Dados Abertos.

O conjunto possui dados dos eventos de internação hospitalar e dos pacientes internados (como sexo, faixa etária e doença). O conjunto inclui, ainda, o “Tempo de Permanência Hospitalar”, que corresponde ao tempo médio, em dias, que o paciente ficou no hospital. Ele é calculado pela ANS, considerando o tempo decorrido entre a data de entrada da paciente na internação hospitalar e a data da saída internação (incluindo motivos como as saídas por alta hospitalar, transferência externa ou óbito).

O “Tempo de Permanência Hospitalar” representa um indicador clássico e relevante para avalição da eficiência da atenção à saúde no âmbito hospitalar. O tempo médio de permanência hospitalar é utilizado para avaliar a eficiência do cuidado em saúde proporcionado aos pacientes, verificando se o tempo para recuperação dos pacientes ocorre em um período considerado adequado. No âmbito do setor suplementar de saúde brasileiro, entende-se como desejável que a média do tempo de internação não ultrapasse 5 dias (https://www.gov.br/ans/pt-br/arquivos/assuntos/prestadores/qualiss-programa-de-qualificacao-dos-prestadores-de-servicos-de-saude-1/Ficha_tecnica_indicador_5.pdf).

## Hipóteses do Problema

As hipóteses que tracei são as seguintes:

1. **Os pacientes com mais idade tendem a ter internações hospitalares com um maior tempo de duração?**

2. **Existe uma correlação entre o sexo do paciente e o tempo de internação hospitalar?**

3. **O tipo de doença do paciente está relacionado com um maior tempo de internação hospitalar?**

## Tipo de Problema

Dado um conjunto de características dos pacientes e da internação, o objetivo é verificar a possibilidade de prever ou classificar o risco da necessidade de internações hospitalares com permanências mais longas, com base nessas características. ## Matriz de Correlação

A matriz de correlação mede a força e a direção de uma relação linear que os atributos numéricos das espécies podem ter. Valores próximos a 1 indicam uma forte correlação positiva, -1 uma forte correlação negativa, e 0 ausência de correlação linear.

Antes de gerar a Matriz de Correlação, a faixa etária foi convertida para numérica, considerando o número de anos inicial da faixa.

```python
# Função para converter a faixa etária para a idade inicial da faixa
def convert_faixa_etaria_to_int(faixa):
    if pd.isna(faixa):
        return np.nan

    faixa_s = str(faixa).lower() # Converter o dado string

    # Checar os dados de faixa etária não identificada ou não informada
    if 'identificad' in faixa_s or 'não informado' in faixa_s or 'nao informado' in faixa_s:
        return np.nan

    if 'ou mais' in faixa_s:
        # Extrair os números dos casos 'ou mais'
        return int(faixa_s.split(' ou mais')[0].strip())
    elif 'a' in faixa_s:
        # Para os intervalos como '50 a 59', obter o primeiro número
        return int(faixa_s.split(' a ')[0].strip())
    return np.nan # Caso haja outros formatos inesperados

# Aplicar a função para criar a nova coluna numérica
df['FAIXA_ETARIA_NUM'] = df['FAIXA_ETARIA'].apply(convert_faixa_etaria_to_int)

print("Coluna 'FAIXA_ETARIA_NUM' criada com sucesso:")
display(df[['FAIXA_ETARIA', 'FAIXA_ETARIA_NUM']].head())
```

```python
# Matriz de correlação
print("\nMatriz de Correlação:")

# Selecionar colunas numéricas para calcular a correlação
# Incluímos 'FAIXA_ETARIA_NUM' criada para análise numérica da faixa etária.
numeric_df = df.select_dtypes(include=np.number)

# Considerar correlação nas colunas mais relevantes para as hipóteses, como 'TEMPO_DE_PERMANENCIA',
# 'FAIXA_ETARIA_NUM', 'QT_DIARIA_ACOMPANHANTE', 'QT_DIARIA_UTI', etc.
correlation_matrix = numeric_df[['TEMPO_DE_PERMANENCIA', 'FAIXA_ETARIA_NUM', 'QT_DIARIA_ACOMPANHANTE', 'QT_DIARIA_UTI']].corr()

print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
# Selecionar colunas numéricas relevantes para o mapa de calor
numeric_cols_for_heatmap = ['TEMPO_DE_PERMANENCIA', 'FAIXA_ETARIA_NUM', 'QT_DIARIA_ACOMPANHANTE', 'QT_DIARIA_UTI']
correlation_matrix_for_plot = df[numeric_cols_for_heatmap].corr()
sns.heatmap(correlation_matrix_for_plot, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação das Características Numéricas Relevantes')
plt.show()
```

O mapa de calor da matriz de correlação revela que a correlação mais relevante identificada se refere à faixa etária dos pacientes. O coeficiente de correlação de proximadamente 0,23 indica uma correlação linear positiva, com uma tendência de aumento na duração da internação hospitalar à medida que a faixa etária dos pacientes aumenta. Contudo, essa relação não é muito forte, possivelmente em decorrência de outros fatores, como o tipo de doença e condições clínicas dos pacientes, bem como de fatores ligados aos cuidados em saúde proporcionados durante a internação, estarem influenciando no tempo de permanência dos pacientes na internação

A média é uma medida de tendência central que representa o valor típico ou o ponto de equilíbrio de um conjunto de dados. É calculada somando-se todos os valores e dividindo-se pelo número total de observações. É sensível a valores extremos (outliers).

### Desvio Padrão

O desvio padrão é uma medida de dispersão que quantifica a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os pontos de dados tendem a estar próximos da média do conjunto, enquanto um desvio padrão alto indica que os pontos de dados estão espalhados por uma faixa maior de valores. Ele é a raiz quadrada da variância.

```python
# desvio padrão dos atributos numéricos do dataset
df_copy.describe().loc['std']
```

## Agrupamentos do Tempo de Permanência `

Para proporcionar uma nova forma de visualização com os dados limpos, foi realizada abaixo a criação da categoria de 'mais de 30 dias'e gerado um novo gráfico para ilustrar a distribuição.

```python
def categorize_permanencia_individual(tempo):
    if tempo > 30:
        return 'Mais de 30 dias'
    else:
        return f'{int(tempo)} dias'

df_copy['TIPO_PERMANENCIA_DETALHADO'] = df_copy['TEMPO_DE_PERMANENCIA'].apply(categorize_permanencia_individual)

print("Distribuição do novo agrupamento 'TIPO_PERMANENCIA_DETALHADO':")
display(df_copy['TIPO_PERMANENCIA_DETALHADO'].value_counts().sort_index())

print("\nPrimeiras 5 linhas do DataFrame com a nova coluna:")
display(df_copy.head())
```

Para criar uma visualização adicional de agrupamentos do TEMPO_DE_PERMANENCIA foi utilizada a função pd.cut, que permite segmentar e classificar os dados em intervalos. Foram definidos os seguintes intervalos e rótulos:
- **1 dia:** Para internações de exatamente 1 dia.
- **2 a 7 dias:** Para internações que duram entre 2 e 7 dias.
- **8 a 30 dias:** Para internações que duram entre 8 e 30 dias.
- **Mais de 30 dias:** Para internações com duração superior a 30 dias.

```python
# Definir os limites dos bins
bins = [0, 1, 7, 30, df['TEMPO_DE_PERMANENCIA'].max() + 1]

# Definir os rótulos para cada bin
labels = ['1 dia', '2 a 7 dias', '8 a 30 dias', 'Mais de 30 dias']

# Criar a nova coluna 'TIPO_PERMANENCIA_BINS' usando pd.cut
df_copy['TIPO_PERMANENCIA_BINS'] = pd.cut(df_copy['TEMPO_DE_PERMANENCIA'], bins=bins, labels=labels, right=True, include_lowest=True)

print("Distribuição do novo agrupamento 'TIPO_PERMANENCIA_BINS':")
display(df_copy['TIPO_PERMANENCIA_BINS'].value_counts().sort_index())

print("\nPrimeiras 5 linhas do DataFrame com a nova coluna 'TIPO_PERMANENCIA_BINS':")
display(df_copy.head())
```

```python
plt.figure(figsize=(8, 6))
# Distribuição do tempo de permanência (agrupado)
sns.histplot(df_copy['TIPO_PERMANENCIA_BINS'], kde=True)
plt.title('Distribuição do Tempo de Permanência')
plt.xlabel('Tempo de Permanência (em dias)')
plt.ylabel('Frequência')
plt.show()
```

A visualização da distribuição do tempo de permanência agrupado evidencia que, mesmo com a criação de categorias a partir da agregação da quantidade de dias em períodos superiores a 1 dia (2 a 7 dias, 8 a 30 dias, e mais de 30 dias), as internações com 1 dia configuram a condição mais frequente do dataset.
